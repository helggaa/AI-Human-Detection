# Model Evaluation

This notebook evaluates the trained EfficientNetV2 model on the test dataset.

Pipeline

1. Load dataset
2. Build model
3. Train model
4. Evaluate model
5. Generate evaluation reports
6. Visualize results

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent

if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))
```

```python
from src.preprocessing import DatasetPreprocessor
from src.model import ModelBuilder
from src.trainer import ModelTrainer
from src.evaluator import ModelEvaluator
from src.visualization import Visualizer
```

```python
preprocessor = DatasetPreprocessor()

dataset = preprocessor.run()
```

```python
builder = ModelBuilder()

artifacts = builder.run()
```

```python
trainer = ModelTrainer(
    builder=builder,
    artifacts=artifacts,
    dataset=dataset,
)

results = trainer.run(
    fine_tune=True,
)
```

```python
evaluator = ModelEvaluator(
    results=results,
    dataset=dataset,
)

evaluation = evaluator.run()
```

```python
evaluation.metrics
```

```python
evaluation.classification_report
```

```python
evaluation.confusion_matrix
```

```python
history = (
    results.fine_tune_history
    if results.fine_tune_history
    else results.initial_history
)

Visualizer.plot_training_history(history)
```

```python
Visualizer.plot_loss_history(history)
```

```python
Visualizer.plot_confusion_matrix(
    evaluation.confusion_matrix,
    evaluator.class_names,
)
```

```python
Visualizer.plot_roc_curve(
    evaluator.labels,
    evaluation.prediction_results.probabilities[:, 1],
)
```

```python
Visualizer.save_training_history(history)

Visualizer.save_confusion_matrix(
    evaluation.confusion_matrix,
    evaluator.class_names,
)

Visualizer.save_roc_curve(
    evaluator.labels,
    evaluation.prediction_results.probabilities[:, 1],
)
```
